```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("../movie_statistic_dataset.csv")
```

```python
dataset = dataset.drop(dataset.index[0:6])
dataset
```

```python
coloums_target = ["movie_title", "genres", "runtime_minutes", "director_name", "director_deathYear"]
filtered_dataset = dataset[coloums_target].copy()
filtered_dataset
```

```python
filtered_dataset["director_deathYear"] = filtered_dataset["director_deathYear"].replace({"alive": 2, "-" : 0})
filtered_dataset.dropna(inplace=True)
filtered_dataset.reset_index(drop=True, inplace=True)
```

```python
filtered_dataset.to_csv("filtered_dataset.csv")
```
